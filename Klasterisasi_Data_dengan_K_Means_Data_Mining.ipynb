# **Pendahuluan**

Klasterisasi adalah teknik dalam Data Mining yang bertujuan untuk mengelompokkan data berdasarkan kesamaan karakteristiknya. Salah satu algoritma yang sering digunakan adalah K-Means, yang bekerja dengan mengelompokkan data ke dalam K klaster berdasarkan jarak terdekat dengan centroid.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

# **Membuat Dataset Simulasi**

Kita akan menggunakan data sintetis dengan tiga klaster:

```python
# Membuat dataset simulasi dengan 3 klaster
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Visualisasi data sebelum klasterisasi
plt.scatter(X[:, 0], X[:, 1], s=50, alpha=0.5)
plt.title("Data Awal Tanpa Klaster")
plt.show()
```

# **Menerapkan Algoritma K-Means**

```python
# Menjalankan algoritma K-Means dengan K=3
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Mendapatkan label klaster
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

# **Visualisasi Hasil Klasterisasi**

```python
# Plot hasil klasterisasi
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("Hasil Klasterisasi dengan K-Means")
plt.legend()
plt.show()
```

# **Evaluasi Hasil Klasterisasi**

Kita dapat mengevaluasi hasil klasterisasi dengan metode Inertia (SSE - Sum of Squared Errors):

```python
print(f'Inertia: {kmeans.inertia_}')
```

Selain itu, kita bisa menentukan jumlah klaster optimal menggunakan metode Elbow Method:

```python
inertia = []
k_values = range(1, 10)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(k_values, inertia, marker='o')
plt.xlabel('Jumlah Klaster')
plt.ylabel('Inertia')
plt.title('Metode Elbow untuk Menentukan K Optimal')
plt.show()
```
